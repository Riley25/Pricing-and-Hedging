```r
suppressMessages(library('xlsx'))
suppressMessages(library('dplyr'))
suppressMessages(library('derivmkts'))
suppressMessages(library('ggplot2'))

# This is a setting to change the size of plotting.
options(repr.plot.width = 5, repr.plot.height =5)
```

```r
par(mfrow =c(1,1))

s = 30       # Stock Price at Time = 0
k = 31       # Strike Price
v = .30      # Volatility
r = .025      # risk-free rate
t = (6/12)   # time to maturity (in years)
d = 0        # dividend yeild
nsteps = 50  # number of time steps


derivmkts::binomopt(s, k , v , r , t , d , nsteps,
                    putopt=FALSE, american=FALSE,
                    returntrees=TRUE)
```

```r
s = 90       # Stock Price at Time = 0
k = 93       # Strike Price
v = .28      # Volatility
r = .03      # risk-free rate
t = (9/12)   # time to maturity (in years)
d = 0        # dividend yeild
nsteps = 3   # number of time steps


df = derivmkts::binomopt(s, k , v , r , t , d , nsteps,
                    putopt=FALSE, american=TRUE,
                    returntrees=TRUE)


c = as.numeric(df$price)
print(c)
```

```r
#Data Prep and Greeks
#r = 0.03
#dividend = 0
s = 1213.8599 #Since Snapshot Date is the same, Close Price on 11/03/2021 is taken as Spot
```

```r
df_calls <- df[(df$TYPE == "CALL"),]

call_greeks = greeks(bscall(s, df_calls$Strike, df_calls$IV, 0.03, df_calls$Years_to_Exp_Date, 0)
            , complete=TRUE, long=FALSE, initcaps=TRUE)

df_calls$Delta <- call_greeks$Delta
df_calls$Gamma <- call_greeks$Gamma
df_calls$Vega <- call_greeks$Vega
df_calls$Rho <- call_greeks$Rho
df_calls$Theta <- call_greeks$Theta
```

```r
df_puts <- df[(df$TYPE == "PUT"),]
put_greeks = greeks(bsput(s, df_puts$Strike, df_puts$IV, 0.03, df_puts$Years_to_Exp_Date, 0)
                  , complete=TRUE, long=FALSE, initcaps=TRUE)

df_puts$Delta <- put_greeks$Delta
df_puts$Gamma <- put_greeks$Gamma
df_puts$Vega <- put_greeks$Vega
df_puts$Rho <- put_greeks$Rho
df_puts$Theta <- put_greeks$Theta
```

```r
## Data Viz

df_11_5_C <- df_calls[(df_calls$Exp_date == "11/5/2021"),]
df_11_5_P <- df_puts[(df_puts$Exp_date == "11/5/2021"),]



df_list = list(df_11_5_C,df_11_5_P)




#Strike v Price 

for(x in 1:length(df_list)){
  
  print(ggplot(df_list[[x]], aes(x=Strike,y=Last)) + geom_line() +
          labs(title = paste0("Strike vs Price for"
                              ,unique(df_list[[x]]$Exp_date),"",unique(df_list[[x]]$TYPE))))  
  
  
}


ggplot(df_11_5_C, aes(x=Delta)) +
  geom_line(aes(y=Gamma)) +
  
  labs(x = "Delta", 
      y = "Gamma")


ggplot(df_11_5_P, aes(x=Strike)) +
  geom_line(aes(y=Delta)) +
  
  labs(x = "Strike", 
      
       y = "Delta")

ggplot(df_calls, aes(x=Years_to_Exp_Date)) +
  geom_point(aes(y=Theta)) +
  labs(x = "Years_to_exp_date", 
       y = "Theta")


ggplot(df_calls[(df_calls$Strike == 1250),], aes(x=Years_to_Exp_Date)) +
  geom_line(aes(y=Last)) +
  labs(x = "Years_to_exp_date", 
       y = "Last",
       title = "Years to Exp vs Option Price for $1250C")
```
